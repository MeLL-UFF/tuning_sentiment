```python
def montaTabelaLatex(df, metrica):
  first = True
  for ind in df.index:
    if first == False:
      print("\\rule{0pt}{3.0ex}")
    else:
      first = False
    print (str(df["Data_Set"][ind])+'&'+str(df[metrica][ind])+'&'+str(df["Classifier_Model"][ind])+'&'+str(df["Embedding"][ind])+'\\\\')
```

```python
def monta_tabela_latex_top_rank_sum(df_acc, df_f1):
  first = True
  for ind in df_acc.index:
    if first == False:
      print("\\rule{0pt}{3.0ex}")
    else:
      first = False
    print (str(df_acc["Embedding"][ind])+'&'+str(df_acc["Classifier_Model"][ind])+'&' \
           +str(round(df_acc['rank'][ind]/22,2))+'&'+str(df_f1["Embedding"][ind])+'&' \
           +str(df_f1["Classifier_Model"][ind])+'&'+str(round(df_f1['rank'][ind]/22,2))+ \
           '\\\\')
           
```

```python
def monta_tabela_latex_top_rank_sum_class(df_acc, df_f1,num_emeb):
  first = True
  for ind in df_acc.index:
    if first == False:
      print("\\rule{0pt}{3.0ex}")
    else:
      first = False
    print (str(df_acc["Classifier_Model"][ind])+'&' \
           +str(round(df_acc['rank'][ind]/(num_emeb*22),2))+'&' \
           +str(df_f1["Classifier_Model"][ind])+'&'+str(round(df_f1['rank'][ind]/(num_emeb*22),2))+ \
           '\\\\')
```

```python
def monta_tabela_latex_top_rank_sum_emb(df_acc, df_f1,num_class):
  first = True
  for ind in df_acc.index:
    if first == False:
      print("\\rule{0pt}{3.0ex}")
    else:
      first = False
    print (str(df_acc["Embedding"][ind])+'&' \
           +str(round(df_acc['rank'][ind]/(22*num_class),2))+'&'+str(df_f1["Embedding"][ind])+'&' \
           +str(round(df_f1['rank'][ind]/(22*num_class),2))+ \
           '\\\\')
```
