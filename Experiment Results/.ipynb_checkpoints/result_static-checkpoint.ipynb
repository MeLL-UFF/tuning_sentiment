```python
def deParaEmbtName(textDe):
  if textDe == 'Roberta: roberta-base / STATIC_AVG':
    return 'Robert static'
  elif textDe == 'Bert: bert-base-uncased / STATIC_AVG':
    return 'Bert static'
  elif textDe == 'Bertweet: vinai/bertweet-base / STATIC_AVG':
    return 'BERTweet static'
  else:
    return textDe
```

```python
def deParaDatasetName(textDe):
  if textDe == 'ntua':
    return 'ntu'
  elif textDe == 'SemEval15-Task11':
    return 'S15'
  elif textDe == 'iphone':
    return 'iph'
  elif textDe == 'Narr-KDML-2012':
    return 'Nar'
  elif textDe == 'VADER':
    return 'Vad'
  elif textDe == 'SemEval17-test':
    return 'S17'
  elif textDe == 'debate08':
    return 'OMD'
  elif textDe == 'irony':
    return 'iro'
  elif textDe == 'sarcasm':
    return 'sar'
  elif textDe == 'sentiment140':
    return 'stm'
  elif textDe == 'person':
    return 'per'
  elif textDe == 'hobbit':
    return 'hob'
  elif textDe == 'movie':
    return 'mov'
  elif textDe == 'sanders':
    return 'san'
  elif textDe == 'archeage':
    return 'arc'
  elif textDe == 'SemEval18':
    return 'S18'
  elif textDe == 'STS-gold':
    return 'STS'
  elif textDe == 'SentiStrength':
    return 'SSt'
  elif textDe == 'Target-dependent':
    return 'Tar'
  elif textDe == 'SemEval13':
    return 'S13'
  elif textDe == 'SemEval16':
    return 'S16'
  else:
    return textDe
```

```python
def get_result_resume(df):
    df_t = df.copy()
    df_t.index=df_t['Embedding']
    del df_t['Embedding']
    df_t = df_t.T
    df_rank = pd.DataFrame()
    for dts in df['Embedding']:
        rank = pd.DataFrame(df_t[dts].rank(ascending=False))
        df_rank = pd.concat([df_rank,rank.T],axis=0)

    dict_emb_sum = pd.DataFrame(df_rank.sum(),columns=['ranksum'])
    dict_emb_sum = dict_emb_sum.reindex([u'Bert: bert-base-uncased / STATIC_AVG', u'DeepMoji', u'EWE',
           u'Emo2Vec', u'FastText', u'GloveTW', u'GloveWP',
           u'Roberta: roberta-base / STATIC_AVG', u'SSWE', u'TFIDF', u'W2V-GN',
           u'W2VAraque', u'W2VEdin', u'Bertweet: vinai/bertweet-base / STATIC_AVG'])

    dict_emb_ranksum = dict_emb_sum.rank(ascending=True)
    dict_emb_ranksum.columns=['position']
    dict_emb_ranksum = dict_emb_ranksum.reindex([u'Bert: bert-base-uncased / STATIC_AVG', u'DeepMoji', u'EWE',
           u'Emo2Vec', u'FastText', u'GloveTW', u'GloveWP',
           u'Roberta: roberta-base / STATIC_AVG', u'SSWE', u'TFIDF', u'W2V-GN',
           u'W2VAraque', u'W2VEdin', u'Bertweet: vinai/bertweet-base / STATIC_AVG'])
    sum_wins=[]
    for col in df_rank.columns:
        sum_wins.append(len(df_rank[df_rank[col]<=1.5][col]))

    sum_wins = pd.DataFrame(sum_wins,index=df_rank.columns,columns=['wins'])
    sum_wins = sum_wins.reindex([u'Bert: bert-base-uncased / STATIC_AVG', u'DeepMoji', u'EWE',
           u'Emo2Vec', u'FastText', u'GloveTW', u'GloveWP',
           u'Roberta: roberta-base / STATIC_AVG', u'SSWE', u'TFIDF', u'W2V-GN',
           u'W2VAraque', u'W2VEdin', u'Bertweet: vinai/bertweet-base / STATIC_AVG'])

    final_result = pd.concat([sum_wins,dict_emb_sum,dict_emb_ranksum],axis=1)
    return final_result

```

```python
def isTheWinner(df,model,dataset):
    df_t = df.copy()
    df_t.index=df_t['Embedding']
    del df_t['Embedding']
    df_t = df_t.T
    if df_t[dataset].rank(ascending=False)[model]==1:
        return True
    else:
        return False
```

```python
def printWinnerTableFormat(df,embVal,emb, dsname):
  if isTheWinner(df,emb, dsname)==True:
    return '\\textbf{'+str(embVal)+'}'
  return embVal
```
